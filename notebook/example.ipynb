# Frequency domain

```python
hp, hc = get_fd_waveform(approximant='SEOBNRv5E',
                                 mass1=30,
                                 mass2=30,
                                 spin1z=0.1,
                                 f_lower=20,
                                 delta_f = 1/64,
                                 eccentricity=0.2,rel_anomaly=0)
```

```python
for e in [0,0.1,0.2]:
    hp, hc = get_td_waveform(approximant='SEOBNRv5E',
                                 mass1=30,
                                 mass2=30,
                                 spin1z=0.1,
                                 delta_t=1.0/4096,
                                 f_lower=20,
                                 eccentricity=e,rel_anomaly=0)
    plt.plot(hp.sample_times,hp,label='e='+str(e))
```

```python
for r in [0,np.pi/2,np.pi,2*np.pi,10]:
    hp, hc = get_td_waveform(approximant='SEOBNRv5E',
                                 mass1=30,
                                 mass2=30,
                                 spin1z=0.1,
                                 delta_t=1.0/4096,
                                 f_lower=20,
                                 eccentricity=0.1,rel_anomaly=r)
    plt.plot(hp.sample_times,hp,label='rel_anomaly='+str(r))
plt.legend()
```

```python
for r in [0,2*np.pi]:
    hp, hc = get_td_waveform(approximant='SEOBNRv5E',
                                 mass1=30,
                                 mass2=30,
                                 spin1z=0.1,
                                 delta_t=1.0/4096,
                                 f_lower=20,
                                 eccentricity=0.1,rel_anomaly=r)
    plt.plot(hp.sample_times,hp,label='rel_anomaly='+str(r))
plt.legend()
```
