```python
import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")

from pycbc import waveform

from matplotlib import pyplot as plt
```

```python
hp, hc = waveform.get_td_waveform(approximant="SEOBNRv4_opt",
                                    mass1=10,
                                    mass2=10,
                                    delta_t=1.0/4096,
                                    f_lower=30)
```

```python
hp.plot()
```

```python
hp.start_time
```

```python
hp._epoch
```

```python
hf_shift = hp.to_frequencyseries().cyclic_time_shift(hp.start_time)
ht_shift = hf_shift.to_timeseries()
```

```python
ht_shift.plot()
```

```python
ht_shift.start_time
```

```python
for app in ['SEOBNRv4_ROM','SEOBNRv5E']:
    hp, hc = waveform.get_fd_waveform(approximant=app,
                                    mass1=10,
                                    mass2=10,
                                    delta_f=1.0/16,
                                    f_lower=30)
    plt.figure()
    print(hp.start_time)
    print(hp._epoch)
    ht = hp.to_timeseries()
    ht.plot()
```
